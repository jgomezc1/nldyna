# Linear dynamic analysis of a 2D-frame.

El delta de tiempo especificado en los parametros iniciales del análisis de $0.02 \sec$, no fue lo suficientemente pequeño para obtener una solución precisa del sistema. Cabe señalar que aún cuando en este problema, todos los elementos fueron elasticos-lineales, en un problema no-lineal que considere el posible cambio de la rigidez de los materiales, puede ser requerido un delta de tiempo aún menor.
